### Import Libraries

```python
import pandas as pd #importing libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import Dataset

datasource : https://chronicdata.cdc.gov/Heart-Disease-Stroke-Prevention/National-Vital-Statistics-System-NVSS-National-Car/kztq-p2jf

```python
stroke = pd.read_csv('./datasets/stroke.csv') #importing the dataset
```

```python
stroke.head()
```

### Distribution

```python
sns.set() # for style
 
ax = sns.distplot(stroke["Data_Value"],label="Heart Disease Data");
ax.set(xlabel = "Data Value"); #labelling the axis
plt.title("Data Value"); # for histogram title
plt.legend(); # for label
```

Data Value is positively Skewed and it means that its Mean will be greater than its Median

```python
sns.set() # for style
 
ax = sns.distplot(stroke["AvgConfidenceLimit"],label="Heart Disease Data");
ax.set(xlabel = "Confidence Limit"); #labelling the axis
plt.title("Confidence Limit"); # for histogram title
plt.legend(); # for label
```

The Avg Confidence Limit is Negatively Skewed

```python
sns.set() # for style
 
ax = sns.distplot(stroke["HighConfidenceLimit"],label="Heart Disease Data");
ax.set(xlabel = "High Confidence Limit"); #labelling the axis
plt.title("High Confidence Limit"); # for histogram title
plt.legend(); # for label
```

The High Confidence Limit is positively skewed

```python
sns.set() # for style
 
ax = sns.distplot(stroke["LowConfidenceLimit"],label="Heart Disease Data");
ax.set(xlabel = "Low Confidence Limit"); #labelling the axis
plt.title("Low Confidence Limit"); # for histogram title
plt.legend(); # for label
```

The Low Confidence Limit is positively skewed


### Scatter Plots

```python
plt.figure(figsize=(12,8))
ax = sns.scatterplot(x= 'LowConfidenceLimit', #creating a scatter plot
                y= 'HighConfidenceLimit',
                 hue='Topic',
                data=stroke)
ax.set(xlabel="Low Confidence Limit", ylabel = "High Confidence Limit") #labelling the axis
```

In this scatter plot we can see the trend of different diseases in terms of High and Low confidence Limit from Stroke to Heart Failure. Each Disease values are described in a different color

```python
plt.figure(figsize=(12,8))
ax = sns.scatterplot(x= 'LowConfidenceLimit', #creating a scatter plot
                y= 'HighConfidenceLimit',
                 hue='Break_Out_Category',
                data=stroke)
ax.set(xlabel="Low Confidence Limit", ylabel = "High Confidence Limit") #labelling the axis
```

In this scatter plot we can see the trend of different Categories in terms of High and Low confidence Limit from Age to Gender. Each Category values are described in a different color

```python
ax = sns.scatterplot(x= 'Data_Value', #creating a scatter plot
                y= 'HighConfidenceLimit',
                 hue='Topic_Stroke',
                data=stroke)
ax.set(xlabel="Data Value", ylabel = "High Confidence Limit") #labelling the axis
```

In this scatter plot we can see the trend of Stroke Disease in terms of Data Value and High confidence Limit . "0" means that the person does not have a Stroke and "1" means He/She has it.

```python
plt.figure(figsize=(12,8))
ax = sns.scatterplot(x= 'Data_Value', #creating a scatter plot
                y= 'HighConfidenceLimit',
                 hue='Topic',
                data=stroke)
ax.set(xlabel="Data Value", ylabel = "High Confidence Limit") #labelling the axis
```

In this scatter plot we can see the trend of different Diseases in terms of Data Value and High confidence Limit from Stroke to Heart Failure. Each Disease values are described in a different color

### Correlation

```python
plt.figure(figsize=(12,12)) #fixing the chart size
sns.heatmap(stroke[['Data_Value','AvgConfidenceLimit','HighConfidenceLimit','LowConfidenceLimit','Break_Out_Category_Age','Break_Out_Category_Gender','Break_Out_Category_Overall','Break_Out_Category_Race']].corr(),annot=True, cmap="RdBu_r");
#this heatmap explains the correlation between the variables
```
